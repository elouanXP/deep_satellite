```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import ndimage
```

```python
img1 = plt.imread('before2.jpg')
img2 = plt.imread('after2.jpg')

plt.figure(figsize=(8,8))
plt.imshow(img1)
plt.show()

plt.figure(figsize=(8,8))
plt.imshow(img2)
plt.show()

print(img1.shape)
print(img2.shape)
```

```python
img1 = plt.imread('before2.jpg')
img2 = plt.imread('after2.jpg')

img1 = img1[:,:,0]
img2 = img2[:,:,0]

plt.figure(figsize=(5,5))
plt.imshow(img1,cmap='gray')
plt.show()

plt.figure(figsize=(5,5))
plt.imshow(img2,cmap='gray')
plt.show()

print(img1.shape)
print(img2.shape)
```

```python
img1_copy = np.copy(img1)
img2_copy = np.copy(img2)
plt.hist(img1_copy.ravel(),bins=255)
plt.hist(img2_copy.ravel(),bins=255, alpha=0.8)
plt.show()
```

```python
image2 = img2>130
plt.imshow(image2)
plt.show()
```

```python
image_closing=ndimage.binary_closing(image2, iterations=4)
plt.imshow(image_closing)
plt.show()
```

```python
image_opening=ndimage.binary_opening(image_closing, iterations=5)
plt.imshow(image_opening)
plt.show()
```

```python
plt.figure(figsize=(8,8))
img1 = plt.imread('before2.jpg')
plt.imshow(img1)
plt.imshow(image_opening, alpha=0.35)
plt.show()

plt.figure(figsize=(8,8))
img2 = plt.imread('after2.jpg')
plt.imshow(img2)
plt.imshow(image_opening, alpha=0.35)
plt.show()
```

```python
#number of houses destroyed
label_image2, n_labels = ndimage.label(image_opening)
n_houses = 116 #change the value
print(n_labels)
print(np.abs(n_labels-n_houses)/max(n_houses,n_labels)*100,'% error')
plt.imshow(label_image2)
plt.show()
```
